## Problem 2 [30 Points]
The **Central Limit Theorem** states that the sum of $N$ independent and identically distributed random varaiables converges a normal distribution as $N$ increases. In this problem, we will test this theorem through a numerical experiment.

(a) Create a vector $v1$ that contains $10^6$ samples that are uniformly distributed between $[-0.5, 0.5]$, Plot the histogram with $40$ bins of this vector (This is the case of N=1).

(b) Create another vector $v2$ that contain $10^6$ samples that are uniformly distributed between $[-0.5, 0.5]$. Plot the histogram with $40$ bins of the sum of $v1$ and $v2$ (This is the case of $N=2$).

(c) Repeat (b) for $N = 3$ and $6$ and comment on what you see.

```python
v1 = -0.5+np.random.rand(100000)*(0.5-(-0.5))
plt.figure(figsize=(4, 3))
plt.hist(v1,bins = 40,label = "N=1")
plt.xlabel('v1')
plt.ylabel('frequency')
plt.title("histogram of uniformly distributed samples (N=1)")
plt.legend()
plt.show()
```

```python
v2 = -0.5+np.random.rand(100000)*(0.5-(-0.5))
plt.figure(figsize=(4, 3))
plt.hist(v1+v2,bins = 40,label = "N=2")
plt.xlabel('v1+v2')
plt.ylabel('frequency')
plt.title("histogram of sum of two uniformly distributed samples (N=2)")
plt.legend()
plt.show()
```

```python
v3 = -0.5+np.random.rand(100000)*(0.5-(-0.5))
plt.figure(figsize=(4, 3))
plt.hist(v1+v2+v3,bins = 40,label = "N=3")
plt.xlabel('v1+v2+v3')
plt.ylabel('frequency')
plt.title("histogram of sum of three uniformly distributed samples (N=3)")
plt.legend()
plt.show()
```

```python
v4 = -0.5+np.random.rand(100000)*(0.5-(-0.5))
v5 = -0.5+np.random.rand(100000)*(0.5-(-0.5))
v6 = -0.5+np.random.rand(100000)*(0.5-(-0.5))

plt.figure(figsize=(4, 3))
plt.hist(v1+v2+v3+v4+v5+v6,bins = 40,label = "N=6")
plt.xlabel('v1+v2+v3+v4+v5+v6')
plt.ylabel('frequency')
plt.title("histogram of sum of six uniformly distributed samples (N=6)")
plt.legend()
plt.show()
```

For N=1 (Single Uniform Distribution), the histogram is uniformly distributed between [-0.5, 0.5], as expected.

For N=2 (Sum of Two Uniform Distributions), the histogram starts forming a triangular shape, as the sum of two uniform distributions creates a piecewise linear distribution.

For N=3 (Sum of Three Uniform Distributions), the histogram is smoother, showing an early resemblance to a normal distribution.

For N=6 (Sum of Six Uniform Distributions), the histogram is now nearly Gaussian, confirming that as N increases, the sum of independent uniform distributions approaches a normal distribution.

This experiment visually confirms the **Central Limit Theorem**—as we sum more independent and identically distributed (i.i.d.) random variables, their distribution approaches a normal distribution.

## Problem 3 [30 Points]

Matrix Representation of Fibonacci Sequence:

Recall Fibonacci sequence start with 0, 1 and the following $F_n = F_{n-1}+F_{n-2}$ for $n\geq3$. One can use a matrix representation of Fibonacci Sequence:
$$\underbrace{\begin{bmatrix}
F_n\\
F_{n-1}
\end{bmatrix}}_{M_n} = \begin{bmatrix}
1&1\\
1&0
\end{bmatrix}\underbrace{\begin{bmatrix}
F_{n-1}\\
F_{n-2}
\end{bmatrix}}_{M_{n-1}} = \begin{bmatrix}
F_{n-1}+F_{n-2}\\
F_{n-1}
\end{bmatrix}$$

If we denote $M_n = \begin{bmatrix}
F_n\\
F_{n-1}
\end{bmatrix}$, then above becomes

$$M_n = \begin{bmatrix}
1&1\\
1&0
\end{bmatrix}M_{n-1} =  \begin{bmatrix}
1&1\\
1&0
\end{bmatrix}^2 M_{n-2}= \cdots = \begin{bmatrix}
1&1\\
1&0
\end{bmatrix}^{n-1} M_1 = \begin{bmatrix}
1&1\\
1&0
\end{bmatrix}^{n-1} \begin{bmatrix}
1\\
0
\end{bmatrix}$$

Define a function `fibonacci_M(n)` to compute the $n$th Fibonacci vector $M_n$ and output $F_n$. Then create a numpy array `Fib_array` which contains the first 20 Fibonacci number getting from using `fibonacci_M(n)`.

```python
def fibonacci_M(n):
  """Generate the nth Fibonacci vector M_n, output the nth Fibonacci number M_n[0]"""
  A = np.array([[1,1],[1,0]])
  M = np.array([1,0])
  for i in range(n-1):
    M = np.dot(A,M)
  return M[0]

Fib_array = np.zeros(20,dtype = int)
for i in range(1,20):
  Fib_array[i] = fibonacci_M(i)

print(Fib_array)
```
